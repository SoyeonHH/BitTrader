# 2. 데이터 불러오기

sample_id : 개별 샘플의 인덱스

time : x_df는 0분 ~ 1379분, y_df는 0분 ~ 119분의 값을 갖습니다. 동일한 샘플 내 시간 정보

coin_index : 10가지 종류의 코인에 대한 비식별화 인덱스 (0 ~9)

open : open price

high : high price

low : low price

close : close price

volume : 거래량

quote_av : quote asset volume

trades : 거래 건 수

tb_base_av : taker buy base asset volume

tb_quote_av : taker buy quote asset volume

```python
data_path = './open'
train_x_df = pd.read_csv(data_path  + "/train_x_df.csv")
train_y_df = pd.read_csv(data_path  + "/train_y_df.csv")
test_x_df = pd.read_csv(data_path  + "/test_x_df.csv")
```

```python
train_x_df
```

```python
def df_2_3D(df):
    # 3차원 배열로 변환
    sample_size = len(df.sample_id.value_counts())
    time_size = len(df.time.value_counts())
    feature_size = df.iloc[:,2:].shape[1]
    
    array_3d = df.iloc[:,2:].values.reshape(sample_size, time_size, feature_size)
    return array_3d
```

```python
train_x_3d = df_2_3D(train_x_df)
train_y_3d = df_2_3D(train_y_df)
test_x_3d = df_2_3D(test_x_df)
```
